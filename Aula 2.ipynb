```python

#maquina de 3 digitos (entre -4 e 4)

# 1.25
# raciocinio -> 1.25 = 0.125 * 10**1
# num = 1.25 * 10**(-1) = 0.125 
# expoente = -1
# truncamento = 0.125
# arredondamento = 0.125

print("\n1.25")
print("truncamento = 0.125")
print("arredondamento = 0.125")

# 10.053
# raciocinio -> 10.053 = 0.10053 * 10**2
# x = 10.053 * 10**(-2) = 0.10053
# expoente = -2
# truncamento: 0.100
# arredondamento: 0.101

print("\n10.053")
print("truncamento = 0.100")
print("arredondamento = 0.101")

# -238.15
# raciocinio -> -238.15 = -0.23815 * 10**3
# num = -238.15 * 10**(-3) = -0.23815 
# expoente = -3
# truncamento: -0.238
# arredondamento:-0.238

print("\n-238.15")
print("truncamento = -0.238")
print("arredondamento = -0.238")
    
# 2.71828
# raciocinio -> 2.71828 = 0.271828 * 10**1
# num = 2.71828 * 10**(-1) = 0.271828
# expoente = -1
# truncamento: 0.271
# arredondamento: 0.272

print("\n2.71828")
print("truncamento = 0.271")
print("arredondamento = 0.272")

# 0.000007
# raciocinio -> 0.000007 = 0.7 * 10**(-5)
# num = 0.000007 * 10**(5) = 0.7
# expoente = 5
# truncamento: overflow
# arredondamento: overflow

print("\n0.000007")
print("truncamento = overflow")
print("arredondamento = overflow")

# 718235.82
# raciocinio -> 718235.82 = 0.71823582 * 10**(6)
# num = 718235.82 * 10**(-6) = 0.71823582
# expoente = -6
# truncamento: underflow
# arredondamento: underflow

print("\n718235.82")
print("truncamento = underflow")
print("arredondamento = underflow")
```
